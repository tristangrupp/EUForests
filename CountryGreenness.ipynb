```python
datalong = data.melt(id_vars='YEAR', value_name="NDVI", var_name="country")
datalong.head(n=40)
```

```python
colormap = alt.Scale(
    domain=[55, 65, 75, 85],
    range=[
        "white",
        "mediumseagreen",
    ],
    type="sqrt",
)
```

```python
alt.data_transformers.enable('json')
```

```python
threshold = pd.DataFrame([{"threshold": 2003}])
threshold
# Vertical line for vaccination year
rule = alt.Chart(threshold).mark_rule(strokeWidth=4).encode(x="threshold:O")


chart = (
   alt.Chart(datalong)
    .mark_rect()
   .encode(
       x=alt.X("YEAR:O", axis=alt.Axis(title=None, ticks=False)),
       y=alt.Y("country:N", axis=alt.Axis(title=None, ticks=False)),
       color=alt.Color("NDVI:Q", sort="ascending", scale=colormap),
       tooltip=["country", "YEAR", "NDVI"],
   )
   .properties(width=500, height=500)
)

chart #+ rule
```
